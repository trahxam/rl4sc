```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.dpi'] = 256
plt.rcParams['text.usetex'] = True
```

```python
chain_name = 'test'

node_df = pd.read_csv(f'vecsupplychains/{chain_name}/nodes.csv')
edge_df = pd.read_csv(f'vecsupplychains/{chain_name}/edges.csv')

fig, ax = plt.subplots()
fig.set_size_inches(8, 4)

for i in range(edge_df.shape[0]):
    sender = edge_df.loc[i,'edge_sender']
    reciever = edge_df.loc[i,'edge_reciever']
    ax.plot([node_df.loc[sender,'node_pos_x'], node_df.loc[reciever,'node_pos_x']], 
            [node_df.loc[sender,'node_pos_y'], node_df.loc[reciever,'node_pos_y']], color='black', linestyle='--', zorder=0)

ax.scatter(node_df['node_pos_x'], node_df['node_pos_y'], ec='black', fc='white', s=400, zorder=10, marker='s')

node_df['node_inv_txt'] = node_df['node_prod_time'].apply(lambda x: str([0]*x))

for i in range(node_df.shape[0]):
    node = node_df.loc[i,'nodes']
    ax.text(node_df.loc[i,'node_pos_x'], node_df.loc[i,'node_pos_y'], node, zorder=20, verticalalignment='center', horizontalalignment='center')

ax.set_xticks([])
ax.set_yticks([])

fig.show()
```
